```python
import matplotlib.pyplot as plt
import json
```

```python
def plot_training_history(history, name = ''):
    epochs = range(1, len(history['train_loss']) + 1)

    plt.figure(figsize=(12, 5))

    # Plot Loss
    plt.subplot(1, 2, 1)
    plt.plot(epochs, history['train_loss'], 'b-', label='Training Loss')
    plt.plot(epochs, history['val_loss'], 'r-', label='Validation Loss')
    plt.title('Loss vs Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.grid()
    plt.legend()

    # Plot Accuracy
    plt.subplot(1, 2, 2)
    plt.plot(epochs, history['train_acc'], 'b-', label='Training Accuracy')
    plt.plot(epochs, history['val_acc'], 'r-', label='Validation Accuracy')
    plt.title('Accuracy vs Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid()
    plt.savefig(f'../images/train_{name}.png')
    plt.show()
```

```python
pathv3 = "../histories/inceptionv3_history.json"
with open(pathv3, 'r') as file:
    inceptionv3_history = json.load(file)

plot_training_history(inceptionv3_history, 'inceptionv3')
```

```python
pathx = "../histories/xception_history.json"
with open(pathx, 'r') as file:
    xception_history = json.load(file)
```

```python
plot_training_history(xception_history, 'xception')
```

```python
def plot_tain_invsx(path1, path2):
    with open(path1, 'r') as file:
        history1 = json.load(file)
    with open(path2, 'r') as file:
        history2 = json.load(file)

    epochs = range(1, len(history1['train_loss']) + 1)

    plt.figure(figsize=(12, 5))

    # Plot Loss
    plt.subplot(1, 2, 1)
    plt.plot(epochs, history1['train_loss'], 'b-', label='InceptionV3 Training Loss')
    plt.plot(epochs, history2['train_loss'], 'r-', label='Xception Training Loss')
    plt.title('Training Loss Comparison')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.grid()
    plt.legend()

    # Plot Accuracy
    plt.subplot(1, 2, 2)
    plt.plot(epochs, history1['train_acc'], 'b-', label='InceptionV3 Training Accuracy')
    plt.plot(epochs, history2['train_acc'], 'r-', label='Xception Training Accuracy')
    plt.title('Training Accuracy Comparison')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid()
    plt.savefig('../images/train_invsx.png')
    plt.show()
```

```python
plot_tain_invsx(pathv3, pathx)
```

```python
def plot_different_activations(path1, path2, path3 = None):
    with open(path1, 'r') as file:
        history1 = json.load(file)
    with open(path2, 'r') as file:
        history2 = json.load(file)
    if path3:
        with open(path3, 'r') as file:
            history3 = json.load(file)

    epochs = range(1, len(history1['train_loss']) + 1)

    plt.figure(figsize=(12, 5))

    # Plot Loss
    plt.subplot(1, 2, 1)
    plt.plot(epochs, history1['train_loss'], 'b-', label='No activation')
    plt.plot(epochs, history2['train_loss'], 'r-', label='ReLU activation')
    plt.plot(epochs, history3['train_loss'], 'g-', label='ELU activation')
    plt.title('Training Loss Comparison')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.grid()
    plt.legend()

    # Plot Accuracy
    plt.subplot(1, 2, 2)
    plt.plot(epochs, history1['train_acc'], 'b-', label='No activation')
    plt.plot(epochs, history2['train_acc'], 'r-', label='ReLU activation')
    plt.plot(epochs, history3['train_acc'], 'g-', label='ELU activation')
    plt.title('Training Accuracy Comparison')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid()
    plt.savefig('../images/train_activations.png')
    plt.show()
```

```python
path1 = "../histories/xception_history.json"
path2 = "../histories/xception_history_relu.json"
path3 = "../histories/xception_history_elu.json"

plot_different_activations(path1, path2, path3)
```
